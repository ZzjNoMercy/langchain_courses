# profile


> 需要langchain>=1.1才支持这个属性。

LangChain的聊天模型可以通过profile属性暴露一个包含所支持功能和能力的字典：

```json
{
  "max_input_tokens": 400000,
  "image_inputs": True,
  "reasoning_output": True,
  "tool_calling": True
  ...
}
```

在 LangChain 中，model.profile并不是硬编码在模型类内部的静态属性，而是动态构建并附加到模型实例上的一个字典（dict）属性。它的数据来源主要有两个部分：
1. 上游数据源：[models.dev项目](https://models.dev/)
2.  LangChain 集成包中的增强数据（augmentations），例如：langchain_openai/data/profile_augmentations.toml

LangChain使用一个CLI工具langchain-model-profiles来动态构建profile：
1. 从 models.dev 拉取最新模型数据；
2. 合并对应集成包中的 profile_augmentations.toml；
3. 生成最终的 profile 数据，并在模型实例初始化时注入为.profile 属性。
