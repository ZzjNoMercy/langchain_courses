中间件（Middleware） 提供了一种更精细地控制代理Agent运行机制的方法。它在以下场景中非常有用：
- 通过日志记录、数据分析和调试来追踪代理行为。
- 转换提示词（Prompts）、工具选择和输出格式。
- 添加重试机制、备用方案（Fallbacks）和提前终止逻辑。
- 应用速率限制、安全护栏（Guardrails）和敏感信息（PII）检测。

如下图所示，LangChain提供了before_agent、before_model等hooks，中间件正是通过在特定的hook对请求进行拦截，从而实现对请求的修改、记录、分析等功能。

# 注册中间件

有以下两种方法进行中间件的注册：
1. 一个元组（tuple）：即使只有一个中间件也要写成 middleware=(middleware,)
2. 一个列表（list）：middleware=[middleware]，一般都用这种方法

```python
from langchain.agents import create_agent
from langchain.agents.middleware import SummarizationMiddleware, HumanInTheLoopMiddleware

agent = create_agent(
    model="gpt-4.1",
    tools=[...],
    middleware=[
        SummarizationMiddleware(...),
        HumanInTheLoopMiddleware(...)
    ],
)
```
